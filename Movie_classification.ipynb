# Data Description

#It appears that you have provided a dataset with various features related to movies. Each row in the dataset represents a different movie, and the columns contain information such as marketing expense, production expense, multiplex coverage, budget, movie length, ratings for lead actors, lead actresses, director, and producer, critic ratings, trailer views, 3D availability, time taken, Twitter hashtags, genre, average age of actors, the number of multiplexes, collection, and whether the movie won a technical Oscar.

#Here's a brief summary:

#Marketing Expense: The amount spent on marketing the movie.

#Production Expense: The cost associated with producing the movie.

#Multiplex Coverage: The percentage of multiplex coverage for the movie.

#Budget: The budget allocated for making the movie.

#Movie Length: The duration of the movie in minutes.

#Lead Actor Rating: The rating of the lead actor.

#Lead Actress Rating: The rating of the lead actress.

#Director Rating: The rating of the director.

#Producer Rating: The rating of the producer.

#Critic Rating: The rating given by critics.

#Trailer Views: The number of views for the movie trailer.

#3D Availability: Whether the movie is available in 3D or not.

#Time Taken: Time taken to make the movie.

#Twitter Hashtags: The number of Twitter hashtags related to the movie.

#Genre: The genre of the movie (e.g., Thriller, Drama, Comedy).

#Average Age of Actors: The average age of actors in the movie.
#Number of Multiplexes: The number of multiplexes showing the movie.

#Collection: The box office collection for the movie.
#Start Tech Oscar: Whether the movie won a technical Oscar (1 for Yes, 0 for No).

```python
from lightgbm import LGBMClassifier
```

```python
lr = LogisticRegression()
dtc = DecisionTreeClassifier()
sv = SVC()
gnb = GaussianNB()
bni = BernoulliNB()
knn = KNeighborsClassifier()
rfc = RandomForestClassifier()
abc = AdaBoostClassifier()
gbc = GradientBoostingClassifier()
rc=RidgeClassifier()
sgd=SGDClassifier()
p=Perceptron()
pac=PassiveAggressiveClassifier()
lgbm=LGBMClassifier()
```

```python
for i in [lr,dtc,sv,gnb,bni,knn,rfc,abc,gbc,gbc,rc,sgd,p,pac,lgbm]:
  print('-------------------------------------',i)
  abc=i.fit(X_train,y_train)
  y_train_pred=abc.predict(X_train)
  y_test_pred=abc.predict(X_test)


  print('Accuracy_Score Train',accuracy_score(y_train_pred,y_train))
  print('-------------------------------------------')
  print('Accuracy_Score Test',accuracy_score(y_test_pred,y_test))
  print('-------------------------------------------')

  print('Classification Train ',classification_report(y_train_pred,y_train))
  print('-------------------------------------------')

  print('Classification Test ',classification_report(y_test_pred,y_test))
  print('-------------------------------------------')

  print('Classification Train ',confusion_matrix(y_train_pred,y_train))
  print('-------------------------------------------')

  print('Classification Test ',confusion_matrix(y_test_pred,y_test))
  print('-------------------------------------------')
```
